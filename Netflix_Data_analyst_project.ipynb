```python
from os import name
df['cast']=df['cast'].fillna('cast not specified')
cast_df=pd.DataFrame()
cast_df=df['cast'].str.split(',',expand=True).stack()
cast_df=cast_df.to_frame()
cast_df.columns=['actor']
actors=cast_df.groupby(['actor']).size().reset_index(name='count')
actors= actors[actors.actor != 'cast not specified']
actors=actors.sort_values(by='count',ascending=False)
top5Actors=actors.head()
top5Actors=top5Actors.sort_values(by='count',)
barchart2=px.bar(top5Actors,x='count',y='actor',title='Top 5 Actors on Netflix')
barchart2
barchart2.show()
```

```python
df1=df[['type','release_year']]
df1=df1.rename(columns={"release_year":"release year","type":"type"})
df2=df1.groupby(['release year','type']).size().reset_index(name='totalcounts')
df2
```

```python
df2=df2[df2['release year']>=2000]
graph= px.line(df2,x='release year',y='totalcounts',color='type',title='Trend of content on netflix')
graph.show()
```

Sentiment Analysis of Netflix content

```python


# Step 1: Take needed columns
df3 = df[['release_year', 'description']].copy()

# Step 2: Create sentiment column properly (NO LOOP MISTAKES)
df3['sentiment'] = df3['description'].astype(str).apply(
    lambda x: 'Positive' if TextBlob(x).sentiment.polarity > 0
    else ('Negative' if TextBlob(x).sentiment.polarity < 0 else 'Neutral')
)

# Step 3: Group data correctly
df_grouped = df3.groupby(['release_year', 'sentiment']).size().reset_index(name='totalcounts')

# Step 4: Filter years
df_grouped = df_grouped[df_grouped['release_year'] >= 2005]

# Step 5: Plot correct graph
fig = px.bar(
    df_grouped,
    x='release_year',
    y='totalcounts',
    color='sentiment',
    title='Sentiment of content on Netflix'
)

fig.show()
```
